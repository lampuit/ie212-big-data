```python
import requests
import pandas as pd
import time
from datetime import datetime, timedelta

# Hm ly token
def get_access_token(api_key, api_secret):
    url = "https://test.api.amadeus.com/v1/security/oauth2/token"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {
        "grant_type": "client_credentials",
        "client_id": api_key,
        "client_secret": api_secret,
    }
    response = requests.post(url, headers=headers, data=data)
    response.raise_for_status()  # Kim tra li
    return response.json()["access_token"]

# Hm gi yu cu API
def fetch_price_metrics(access_token, origin, destination, departure_date, currency, one_way):
    url = "https://test.api.amadeus.com/v1/analytics/itinerary-price-metrics"
    params = {
        "originIataCode": origin,
        "destinationIataCode": destination,
        "departureDate": departure_date,
        "currencyCode": currency,
        "oneWay": str(one_way).lower(),
    }
    headers = {"Authorization": f"Bearer {access_token}"}
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()  # Kim tra li
    return response.json()

# Hm crawl d liu
def crawl_data(api_key, api_secret, origins, destinations, start_date, days, currency, one_way):
    access_token = get_access_token(api_key, api_secret)
    records = []

    for origin in origins:
        for destination in destinations:
            for i in range(days):
                departure_date = (start_date + timedelta(days=i)).strftime("%Y-%m-%d")
                try:
                    data = fetch_price_metrics(access_token, origin, destination, departure_date, currency, one_way)
                    for entry in data.get("data", []):
                        for metric in entry["priceMetrics"]:
                            records.append({
                                "Origin": origin,
                                "Destination": destination,
                                "Departure Date": departure_date,
                                "One-Way": one_way,
                                "Quartile Ranking": metric["quartileRanking"],
                                "Price": float(metric["amount"]),
                            })
                    print(f"Fetched data for {origin} -> {destination} on {departure_date}")
                except Exception as e:
                    print(f"Error fetching data for {origin} -> {destination} on {departure_date}: {e}")
                time.sleep(1)  # Trnh vt gii hn tc  API

    return records

# Lu d liu vo CSV
def save_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")

# Cu hnh API v tham s crawl
API_KEY = "nTPWRgDWc27ZvTRbwA72zPDiKTDldXoZ"
API_SECRET = "YApKI2o2MMqtJ9K0"
ORIGINS = ["MAD"]  # Sn bay khi hnh
DESTINATIONS = ["CDG"]  # Sn bay n
START_DATE = datetime(2024, 1, 1)  # Ngy bt u crawl
DAYS = 3  # S ngy mun crawl (v d: 7 ngy t ngy bt u)
CURRENCY = "EUR"  # Loi tin
ONE_WAY = False  # Chuyn i kh hi

# Chy crawl
data = crawl_data(API_KEY, API_SECRET, ORIGINS, DESTINATIONS, START_DATE, DAYS, CURRENCY, ONE_WAY)

# Lu d liu
save_to_csv(data, "flight_data.csv")
```

```python
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from concurrent.futures import ThreadPoolExecutor
from time import sleep

# Define constants
USER_AGENT = "Mozilla/5.0 (Windows NT 11.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
MAX_WORKERS = 4
TIMEOUT = 20
OUTPUT_DIR = "Processed_Data_PlaneTrip"

# Generate URLs
def generate_urls(places, days_range=5):
    today = datetime.datetime.today().date()
    dates = [today + datetime.timedelta(days=i) for i in range(days_range)]
    urls = [
        f"https://www.traveloka.com/vi-vn/flight/fullsearch?ap={place}&dt={date.strftime('%d-%m-%Y')}.NA&ps=1.0.0&sc=ECONOMY"
        for place in places for date in dates
    ]
    return urls

# Initialize WebDriver
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument(f"user-agent={USER_AGENT}")
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    driver = webdriver.Chrome(options=options)
    return driver

# Crawl a single URL
def crawl_url(driver, url):
    try:
        driver.get(url)
        WebDriverWait(driver, TIMEOUT).until(
            EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'css-1dbjc4n')]")
        ))
        
        # Extract data
        prices = [elem.text for elem in driver.find_elements(By.XPATH, "//h3[contains(@class, 'css-4rbku5')]")]
        brands = [elem.text for elem in driver.find_elements(By.XPATH, "//div[contains(@class, 'css-901oao')]")]
        
        # Add additional data extraction logic as needed
        data = pd.DataFrame({
            "brand": brands,
            "price": prices,
            "url": url
        })
        print(f"Successfully crawled: {url}")
        return data

    except (TimeoutException, NoSuchElementException) as e:
        print(f"Error crawling {url}: {e}")
        return pd.DataFrame()

# Parallel crawl
def parallel_crawl(urls):
    results = []
    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        with init_driver() as driver:
            futures = [executor.submit(crawl_url, driver, url) for url in urls]
            for future in futures:
                try:
                    result = future.result()
                    if not result.empty:
                        results.append(result)
                except Exception as e:
                    print(f"Error in future: {e}")
    return pd.concat(results, ignore_index=True) if results else pd.DataFrame()

# Save data
def save_data(data, filename):
    output_path = f"{OUTPUT_DIR}/{filename}"
    data.to_csv(output_path, index=False)
    print(f"Data saved to {output_path}")

if __name__ == "__main__":
    # Input data
    places = ['SGN.DAD', 'SGN.CXR']

    # Step 1: Generate URLs
    url_list = generate_urls(places)

    # Step 2: Crawl data
    crawled_data = parallel_crawl(url_list)

    # Step 3: Save data
    if not crawled_data.empty:
        today = datetime.datetime.today().strftime('%Y-%m-%d')
        save_data(crawled_data, f"PlaneTrip_{today}.csv")
    else:
        print("No data crawled.")
```
