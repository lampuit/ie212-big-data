```python
from kafka import KafkaProducer
import json
import pandas as pd
import time

# Kafka Configuration
kafka_broker = "127.0.0.1:9092"
kafka_topic = "Item_Outlet_Sales_Prediction"

# Initialize Kafka producer
producer = KafkaProducer(
    bootstrap_servers=kafka_broker,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Load Dataset
dataset_path = "Train.csv"  # Path to the dataset
df = pd.read_csv(dataset_path)

# Ensure consistent columns
required_columns = [
    "Item_Identifier", "Item_Weight", "Item_Visibility", "Item_MRP",
    "Outlet_Establishment_Year", "Item_Fat_Content", "Item_Type",
    "Outlet_Identifier", "Outlet_Size", "Outlet_Location_Type", "Outlet_Type"
]

# Stream Data to Kafka
for _, row in df.iterrows():
    message = {col: row.get(col, "") for col in required_columns}
    producer.send(kafka_topic, value=message)
    print(f"Sent: {message}")
    time.sleep(1)  # Delay for simulation

producer.close()
```
